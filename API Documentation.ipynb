# API Documentation

First import the API and initilize the calss as an object

```python
from API.RateMyProfApi import RateMyProfApi
RMP = RateMyProfApi()
```

## get_schools_by_name("school_name")  
returns a list school object

```python
schools = RMP.get_schools_by_name("The University of Texas")
print(schools)
```

## get_school_by_name("school_name")  
returns a school object

```python
school = RMP.get_school_by_name("The University of Texas at Dallas")
print(school)
```

## get_professors_by_school_and_name(school, "Prof name")  
returns a list of professor objects

```python
Profs = RMP.get_professors_by_school_and_name(school, "smith") 
print(Profs)
```

## get_professor_by_school_and_name(school, "Prof name")  
returns a professor object

```python
Prof = RMP.get_professor_by_school_and_name(school, "Jason Smith") 
print(Prof)
```

## get_ratings("course name")  
returns a list of all ratings asociated with that professor and the course name, if course name is left blank returns all ratings for that professor

```python
print(Prof.get_ratings("CS2337"))
```
